```python
query1 = pd.read_sql_query("SELECT productname, unitsinstock FROM products", con)
query1
```

```python
query2 = pd.read_sql_query("SELECT productname, unitprice FROM products",con)
query2.sort_values(by = ['unitprice'],ascending = False)

# We see that Cote de Blaye is the most expensive at $263.50 and Geitost least expensive at $2.50
```

```python
query3 = pd.read_sql_query("SELECT COUNT(productid),COUNT(discontinued) FROM products", con)
query3
```

```python
query4 = pd.read_sql_query("SELECT products.productname , categories.categoryname FROM products INNER JOIN categories ON products.categoryid = categories.categoryid",con)
query4
```

```python
query5 = pd.read_sql_query("SELECT products.productname, products.unitprice, suppliers.country FROM products INNER JOIN suppliers ON products.supplierid = suppliers.supplierid WHERE suppliers.country NOT LIKE 'USA'",con)
query5
```

```python
query6 = pd.read_sql_query("SELECT SUM(quantity) FROM order_details",con)
query6
```

```python
query7 = pd.read_sql_query("SELECT products.productname , SUM(order_details.quantity) FROM products INNER JOIN order_details ON products.productid = order_details.productid GROUP BY productname",con)
query7
```
